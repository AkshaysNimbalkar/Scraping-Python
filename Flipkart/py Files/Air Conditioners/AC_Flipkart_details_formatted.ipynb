```python
import os
import pandas as pd
import time
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('F:/Test/MX_Scraping/AC_Flipkart_details.csv', header=None)
df.head()
```

```python
df_keys = df.iloc[::2, :]
df_keys.reset_index(drop=True, inplace=True)
df_values = df.iloc[1::2, :]
df_values.reset_index(drop=True, inplace=True)
```

```python
def create_List(df_keys, df_values, specification):
    new_list = ['Info Not Available'] * df_keys.shape[0]
    for i in range(0,df_keys.shape[1]):
        for j in range(0,df_keys.shape[0]):
            if df_keys[i][j] == specification:
                new_list[j] = df_values[i][j]
    return new_list
```
