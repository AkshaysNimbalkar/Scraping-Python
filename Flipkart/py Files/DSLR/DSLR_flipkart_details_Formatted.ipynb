```python
def create_List(df_keys, df_values, specification):
    new_list = ['Info Not Available'] * df_keys.shape[0]
    for i in range(0,df_keys.shape[1]):
        for j in range(0,df_keys.shape[0]):
            if df_keys[i][j] == specification:
                new_list[j] = df_values[i][j]
    return new_list
```
